```python
import sys
sys.path.append('../')

from data import VIDEODataset
import matplotlib.pyplot as plt
import torchvision.transforms as T


def show_images(tensor):
    # Vrifie que le tenseur a bien la forme attendue
    # Convertit les images en format acceptable pour matplotlib (de [C, H, W]  [H, W, C])
    to_pil_image = T.ToPILImage()
    
    # Cre une figure avec une seule ligne et 10 colonnes pour afficher les images
    fig, axs = plt.subplots(1, tensor.shape[0], figsize=(20, 5))
    
    # Parcours et affiche les 10 images
    for i in range(tensor.shape[0]):
        ax = axs[i]                    # Slectionne la sous-figure
        ax.imshow(tensor[i].permute(1,2,0))                 # Affiche l'image
        ax.axis('off')                 # Dsactive les axes

    plt.show()


dhf1k_ = VIDEODataset(
    path="/Users/coconut/Documents/Dataset/Saliency/DHF1K/DHF1K/"
    )

# show output from dataloader
print("Dataloader Salicon" , len(dhf1k_))
for i in range(0 , len(dhf1k_)):
    img, sal, fix, _ = dhf1k_[i]
    break

print(img.shape)
print(sal.shape)
print(fix.shape)
# print(target_size)

show_images(img)
```
